# Dataset 1: USDA (usda_clean_data)

```python
usda = pd.read_csv('../data/silver_layer/usda_clean_data.csv')
```

### `dim_food_usda` — Unique food names

```python
dim_food_usda = (
    usda[['food_name']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_food_usda['FoodID'] = dim_food_usda.index + 1

dim_food_usda = dim_food_usda[[
    'FoodID',
    'food_name'
]]
```

### `dim_category` — Unique food categories

```python

dim_category = (
    usda[['food_category']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_category['CategoryID'] = dim_category.index + 1

dim_category = dim_category[[
    'CategoryID',
    'food_category'
]]
```

### `dim_brand` — Unique brand owner / brand name pairs

```python

dim_brand = (
    usda[['brand_owner', 'brand_name']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_brand['BrandID'] = dim_brand.index + 1

dim_brand = dim_brand[[
    'BrandID',
    'brand_owner',
    'brand_name'
]]
```

### `dim_datatype` — Unique data type values (Branded, SR Legacy, Foundation)

```python

dim_datatype = (
    usda[['data_type']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_datatype['DataTypeID'] = dim_datatype.index + 1

dim_datatype = dim_datatype[[
    'DataTypeID',
    'data_type'
]]
```

### `dim_foodtype_usda` — Unique food type values

```python

dim_foodtype_usda  = (
    usda[['food_type']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_foodtype_usda ['FoodTypeID'] = dim_foodtype_usda .index + 1

dim_foodtype_usda  = dim_foodtype_usda [[
    'FoodTypeID',
    'food_type'
]]
```

### `dim_healthscore` — Unique health score values

```python

dim_healthscore = (
    usda[['health_score']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_healthscore['HealthScoreID'] = dim_healthscore.index + 1

dim_healthscore = dim_healthscore[[
    'HealthScoreID',
    'health_score'
]]
```

### Merge dimension IDs into fact table

```python

fact_foodnutrition = usda.copy()

# Merge FoodID
fact_foodnutrition = fact_foodnutrition.merge(
    dim_food_usda,
    on='food_name',
    how='left'
)

# Merge CategoryID
fact_foodnutrition = fact_foodnutrition.merge(
    dim_category,
    on='food_category',
    how='left'
)

# Merge BrandID
fact_foodnutrition = fact_foodnutrition.merge(
    dim_brand,
    on=['brand_owner', 'brand_name'],
    how='left'
)

# Merge DataTypeID
fact_foodnutrition = fact_foodnutrition.merge(
    dim_datatype,
    on='data_type',
    how='left'
)

# Merge FoodTypeID
fact_foodnutrition = fact_foodnutrition.merge(
    dim_foodtype_usda ,
    on='food_type',
    how='left'
)

# Merge HealthScoreID
fact_foodnutrition = fact_foodnutrition.merge(
    dim_healthscore,
    on='health_score',
    how='left'
)
```

### `fact_foodnutrition` — Select final fact table columns

```python

fact_foodnutrition = fact_foodnutrition[[
    'fdc_id',
    'FoodID',
    'CategoryID',
    'BrandID',
    'DataTypeID',
    'FoodTypeID',
    'HealthScoreID',
    'calories',
    'carbs_g',
    'fat_g',
    'protein_g',
    'sugar_g',
    'fiber_g',
    'sodium_mg',
    'calcium_mg',
    'iron_mg',
    'saturated_fat_g',
    'vitamin_c_mg',
    'cholesterol_mg'
]]
```

# Dataset 3: Allergens (allergens_clean_data)

```python
allergens = pd.read_csv('../data/silver_layer/allergens_clean_data.csv')
```

### `dim_product` — Unique product name / brand pairs

```python
dim_product = (
    allergens[['product_name', 'brands']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_product['ProductID'] = dim_product.index + 1

dim_product = dim_product[[
    'ProductID',
    'product_name',
    'brands'
]]
```

### `dim_nutriscore` — Unique Nutri-Score grades

```python
dim_nutriscore = (
   allergens[['nutriscore_grade']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_nutriscore['NutriScoreID'] = dim_nutriscore.index + 1

dim_nutriscore = dim_nutriscore[[
    'NutriScoreID',
    'nutriscore_grade'
]]
```

### `dim_nova` — Unique NOVA group values

```python
dim_nova = (
    allergens[['nova_group']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_nova['NovaGroupID'] = dim_nova.index + 1

dim_nova = dim_nova[[
    'NovaGroupID',
    'nova_group'
]]
```

### `dim_allergen_type` — Unique allergen combinations

```python
dim_allergen_type = (
    allergens[['allergens']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_allergen_type['AllergenTypeID'] = dim_allergen_type.index + 1

dim_allergen_type = dim_allergen_type[[
    'AllergenTypeID',
    'allergens'
]]
```

### `dim_allergen_cat` — Unique category / subcategory pairs

```python

dim_allergen_cat = (
    allergens[['category', 'subcategory']]
    .drop_duplicates()
    .reset_index(drop=True)
)

dim_allergen_cat['AllergenCatID'] = dim_allergen_cat.index + 1

dim_allergen_cat = dim_allergen_cat[[
    'AllergenCatID',
    'category',
    'subcategory'
]]
```

### Merge dimension IDs into fact table

```python

fact_allergens = allergens.copy()

# Merge ProductID
fact_allergens = fact_allergens.merge(
    dim_product,
    on=['product_name', 'brands'],
    how='left'
)

# Merge NutriScoreID
fact_allergens = fact_allergens.merge(
    dim_nutriscore,
    on='nutriscore_grade',
    how='left'
)

# Merge NovaGroupID
fact_allergens = fact_allergens.merge(
    dim_nova,
    on='nova_group',
    how='left'
)

# Merge AllergenTypeID
fact_allergens = fact_allergens.merge(
    dim_allergen_type,
    on='allergens',
    how='left'
)

# Merge AllergenCatID
fact_allergens = fact_allergens.merge(
    dim_allergen_cat,
    on=['category', 'subcategory'],
    how='left'
)
```

## Cross-Dataset Merge: USDA × Healthy Foods

Joins USDA and Healthy Foods on `food_name` to enable cross-analysis.

- `calories` and `health_score` conflicts are resolved by keeping USDA calories and Healthy Foods health score.
- Duplicated numeric columns (`protein_g`, `fat_g`, etc.) are consolidated using `combine_first`.
- Duplicated ID columns from both fact tables are dropped.

```python
# ── Step 1: inner join on food_name ───────────────────────────────────
usda_labels = (
    fact_foodnutrition
    .merge(dim_food_usda,     on='FoodID',      how='left')
    .merge(dim_category,      on='CategoryID',  how='left')
    .merge(dim_foodtype_usda, on='FoodTypeID',  how='left')
)

merged = usda_labels.merge(healthy_db, on='food_name', how='inner')

# ── Step 2: resolve calories / health_score column conflicts ──────────
# Keep USDA calories (more complete), Healthy Foods health_score (source of truth)
merged = merged.rename(columns={
    'calories_x':     'calories',
    'health_score_y': 'health_score'
})
merged = merged.drop(columns=['calories_y', 'health_score_x'], errors='ignore')

# ── Step 3: consolidate duplicated numeric columns ────────────────────
common_numeric = ['protein_g', 'fat_g', 'carbs_g', 'fiber_g', 'sugar_g', 'sodium_mg']
for col in common_numeric:
    if f'{col}_x' in merged.columns:
        merged[col] = merged[f'{col}_x'].combine_first(merged[f'{col}_y'])
        merged.drop(columns=[f'{col}_x', f'{col}_y'], inplace=True)

# ── Step 4: drop duplicated ID / label columns from the join ──────────
merged.drop(
    columns=['FoodTypeID_x', 'FoodTypeID_y', 'HealthScoreID_x',
             'HealthScoreID_y', 'FoodID_y', 'food_type_y'],
    inplace=True, errors='ignore'
)
merged.rename(columns={'FoodID_x': 'FoodID', 'food_type_x': 'food_type'}, inplace=True)

# ── Step 5: select final columns for analysis ─────────────────────────
merged_final = merged[[
    'food_name', 'food_category', 'food_type',
    'calories', 'health_score',
    'protein_g', 'fat_g', 'carbs_g', 'fiber_g', 'sugar_g', 'sodium_mg'
]]

print('merged_final shape:', merged_final.shape)
print(merged_final.head())
```
